```python
path = '/content/drive/MyDrive/2024_Spring/RBDA/Final_Project/final_joins_csv/case+hosp.csv'
df = pd.read_csv(path)
print(df)
```

```python
df['date'] = pd.to_datetime(df['date'])
df.head(10)
```

```python
# Set the start and end dates
start_date = '2020-01-01'
end_date = '2021-12-31'
# Filter the DataFrame
filtered_df = df[(df['date'] >= start_date) & (df['date'] <= end_date)]
# Display the first few rows of the filtered DataFrame to verify
filtered_df.head(10)
```

```python
# Calculate Pearson correlation and p-value
correlation, p_value = pearsonr(filtered_df['sumConfirmed'], filtered_df['cu_hosp'])

print(f"Correlation Coefficient: {correlation}")
print(f"P-Value: {p_value}")
```
